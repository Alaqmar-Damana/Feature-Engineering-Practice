```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

**Country Code is the common column in both the datasets so now we'll combine both the datasets.**

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
##to check the datatypes
final_df.dtypes
```

##Let's Explore!

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

**In above code we can observe that Zomato is mostly available in India**

```python
#if I just want to get the names of the countries
country_names = final_df.Country.value_counts().index
```

```python
country_names
```

```python
#to get values
country_values = final_df.Country.value_counts().values
```

##PIE CHART

```python
plt.pie(country_values, labels = country_names)
```

```python
#I just want to see the top 3 countries here
plt.pie(country_values[:3], labels=country_names[:3])
```

```python
#if I want to display the percentage
plt.pie(country_values[:3], labels=country_names[:3], autopct="%1.2f%%")
```

###Observation:
Zomato maximum records or transactions are from India then USA then UK

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

#Observations
1: When the rating is between 4.5 to 4.9---> Excellent

2: When the rating is between 4.0 to 4.4---> Very Good

3: When the rating is between 3.5 to 3.9---> Good

4: When the rating is between 3.0 to 3.4---> Average

5: When the rating is between 2.5 to 2.9---> Average

6: When the rating is between 2.0 to 2.4---> Poor


```python
ratings.head()
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
#increase the figure size
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

**Yep! Now I can see the numbers properly**

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue='Rating color')
```

```python
#let's map the colors
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue='Rating color', palette = ['blue','red','orange','yellow','green','green'])
```

##Observation:
  1: Not rated count is very high
  
  2: Maximum number of ratings are in b/w 2.5 to 3.4

##COUNTPLOT

```python
sns.countplot(x="Rating color", data=ratings, palette = ['blue','red','orange','yellow','green','green'])
```
